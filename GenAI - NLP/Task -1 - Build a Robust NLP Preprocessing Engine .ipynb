# Task 1: Conceptual Understanding

1. **"Love" vs "love":** In NLP, these are treated as distinct tokens because computers use ASCII/Unicode values. Standardizing to lowercase ensures the model recognizes them as the same semantic concept.
2. **Stopwords:** If not removed, high-frequency words like "the" or "is" add noise, increase dimensionality, and can skew statistical models (like TF-IDF) toward irrelevant words.
3. **Harmful Stopword Removal:**
   - **Sentiment Analysis:** Removing "not" from "I am not happy" flips the meaning to "I am happy."
   - **Entity Recognition:** Removing "The" from "The Who" or "The Beatles" ruins the proper noun identification.
4. **Stemming vs. Lemmatization:** Stemming is a rule-based approach that chops suffixes (e.g., "running" -> "run"). Lemmatization uses a dictionary (morphological analysis) to find the actual root word (e.g., "better" -> "good").

```python
import re
import pandas as pd
from collections import Counter

def preprocess_text(text):
    """
    Advanced NLP Preprocessing Engine for Task 2 & 7
    """
    # Task 7: Error Handling for non-strings or empty inputs
    if not isinstance(text, str) or not text.strip():
        return []

    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs and Email patterns
    text = re.sub(r'\S+@\S+', '', text) # Emails
    text = re.sub(r'http\S+|www\S+', '', text) # URLs

    # 3. Remove Numbers
    text = re.sub(r'\d+', '', text)

    # 4. Handle repeated characters (e.g., "soooo goooood" -> "so good")
    # This reduces 3 or more repeats to 1
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 5. Remove special characters/emojis (Keep only letters and spaces)
    text = re.sub(r'[^a-z\s]', '', text)

    # 6. Tokenize & Remove extra spaces
    tokens = text.split()

    # 7. Remove short tokens (len <= 2) except "no" and "not"
    keep_list = {'no', 'not'}
    clean_tokens = [t for t in tokens if len(t) > 2 or t in keep_list]

    return clean_tokens
```

```python
def full_pipeline(text_list):
    """
    Task 6: Full Pipeline Implementation
    """
    tokens_list = []
    clean_sentences = []
    
    for txt in text_list:
        processed = preprocess_text(txt)
        tokens_list.append(processed)
        clean_sentences.append(" ".join(processed))
        
    return {
        "tokens": tokens_list,
        "clean_sentences": clean_sentences
    }
```

```python
# Sample Inputs provided in assignment
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Run Pipeline
results = full_pipeline(sample_inputs)

# Task 3: Display Output Format
for i in range(len(sample_inputs)):
    print(f"Original Text: {sample_inputs[i]}")
    print(f"Cleaned Tokens: {results['tokens'][i]}")
    print(f"Cleaned Sentence: {results['clean_sentences'][i]}")
    print("-" * 30)
```

```python
# Task 4: Token Analytics Calculation
analytics_data = []

for i in range(len(sample_inputs)):
    tokens = results['tokens'][i]
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0
    
    analytics_data.append({
        "Sentence Index": i + 1,
        "Total Tokens": total_tokens,
        "Unique Tokens": unique_tokens,
        "Avg Token Length": round(avg_len, 2)
    })

# Display as a Table for better readability
df_analytics = pd.DataFrame(analytics_data)
print(df_analytics)

print("\n--- Analysis Questions ---")
print("1. Which sentence had the most noise? 'Visit https://openai.com now!' (contains URL and punctuation)")
print("2. Which sentence retained the most meaningful tokens? 'I am not happy with this' (Keywords 'not' and 'happy' preserved)")
```

```python
# Task 5: Frequency Analysis
all_tokens = [token for sublist in results['tokens'] for token in sublist]
word_counts = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
for word, count in word_counts.most_common(10):
    print(f"{word}: {count}")

print("\nTop 5 Least Frequent Words:")
# Extract the tail of the most_common list
least_common = word_counts.most_common()[:-6:-1]
for word, count in least_common:
    print(f"{word}: {count}")
```
